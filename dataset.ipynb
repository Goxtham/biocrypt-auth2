```python
import os
import shutil
import random
from math import floor

val_dir = "Val DIR"
train_dir = "Train DIR"

# Ensure reproducibility
random.seed(42)

for class_folder in os.listdir(val_dir):
    val_class_path = os.path.join(val_dir, class_folder)
    
    # Skip non-folders
    if not os.path.isdir(val_class_path):
        continue

    train_class_path = os.path.join(train_dir, class_folder)

    # Create class folder in train if missing
    if not os.path.exists(train_class_path):
        os.makedirs(train_class_path)

    # Get list of image files
    images = [img for img in os.listdir(val_class_path)
              if img.lower().endswith(('.png', '.jpg', '.jpeg'))]

    if len(images) == 0:
        continue

    # Shuffle the images
    random.shuffle(images)

    # Calculate 60%
    num_to_move = floor(len(images) * 0.6)

    # Select first 60%
    images_to_move = images[:num_to_move]

    # Move images
    for img in images_to_move:
        src = os.path.join(val_class_path, img)
        dst = os.path.join(train_class_path, img)
        shutil.move(src, dst)

    print(f"Moved {num_to_move} images from {class_folder} to train.")

print("Finished splitting dataset!")
```
